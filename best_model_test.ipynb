```python
from preproc import PreProcessor
from ensemble_model import EnsembleModel
from bert_model import Model

pp = PreProcessor()
```

```python
import pandas as pd
best_aug = pd.read_csv("aug_datasets/bt.csv")
pp.augmented = best_aug
best_aug = pp.run_preprocess(
    white_func=pp.whitespace_norm,
    punc_func=pp.remove_punc,
    dig_func=pp.remove_digits,
)
best_aug
```

```python
best_model = Model("baseline", layers_to_freeze=5, num_train_epochs=2)
best_model.train(best_aug, pp.raw_dev)
```

```python
pred_outs = best_model.inference(pp.raw_dev[['text']])
```

```python
dev_preds= [pred.argmax(-1) for pred in pred_outs]
dev_preds
```

```python
with open('dev.txt', 'w') as file:
    for pred in dev_preds:
        file.write(str(pred) + '\n')
```
